```python
import numpy as np

# Dados dos voluntários
clara_spo2 = [97, 98, 97, 97, 98, 98]
clara_bpm = [89, 85, 89, 89, 85, 83]

joao_spo2 = [98, 98, 98, 98, 98, 98]
joao_bpm = [83, 85, 83, 85, 83, 89]

solano_spo2 = [98, 98, 98, 98, 98, 98]
solano_bpm = [85, 99, 83, 83, 87, 86]
```

```python
def descriptive_statistics(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    min_val = np.min(data)
    max_val = np.max(data)
    return mean, std_dev, min_val, max_val

def coefficient_of_variation(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    cv = std_dev / mean
    return cv

def correlation(data1, data2):
    return np.corrcoef(data1, data2)[0, 1]

# Análise Descritiva
clara_spo2_stats = descriptive_statistics(clara_spo2)
clara_bpm_stats = descriptive_statistics(clara_bpm)

joao_spo2_stats = descriptive_statistics(joao_spo2)
joao_bpm_stats = descriptive_statistics(joao_bpm)

solano_spo2_stats = descriptive_statistics(solano_spo2)
solano_bpm_stats = descriptive_statistics(solano_bpm)

# Coeficiente de Variação
clara_spo2_cv = coefficient_of_variation(clara_spo2)
clara_bpm_cv = coefficient_of_variation(clara_bpm)

joao_spo2_cv = coefficient_of_variation(joao_spo2)
joao_bpm_cv = coefficient_of_variation(joao_bpm)

solano_spo2_cv = coefficient_of_variation(solano_spo2)
solano_bpm_cv = coefficient_of_variation(solano_bpm)


# Resultados
print("Clara - SpO2: Mean = {:.2f}, Std Dev = {:.2f}, Min = {}, Max = {}".format(*clara_spo2_stats))
print("Joao - SpO2: Mean = {:.2f}, Std Dev = {:.2f}, Min = {}, Max = {}".format(*joao_spo2_stats))
print("Solano - SpO2: Mean = {:.2f}, Std Dev = {:.2f}, Min = {}, Max = {}".format(*solano_spo2_stats))
print(" ")


print("Clara - BPM: Mean = {:.2f}, Std Dev = {:.2f}, Min = {}, Max = {}".format(*clara_bpm_stats))
print("Joao - BPM: Mean = {:.2f}, Std Dev = {:.2f}, Min = {}, Max = {}".format(*joao_bpm_stats))
print("Solano - BPM: Mean = {:.2f}, Std Dev = {:.2f}, Min = {}, Max = {}".format(*solano_bpm_stats))
print(" ")


print("Clara - SpO2 CV: {:.2f}".format(clara_spo2_cv))
print("Joao - SpO2 CV: {:.2f}".format(joao_spo2_cv))
print("Solano - SpO2 CV: {:.2f}".format(solano_spo2_cv))
print(" ")


print("Clara - BPM CV: {:.2f}".format(clara_bpm_cv))
print("Joao - BPM CV: {:.2f}".format(joao_bpm_cv))
print("Solano - BPM CV: {:.2f}".format(solano_bpm_cv))
```
